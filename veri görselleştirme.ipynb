```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.plot(x,y)
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.grid()
plt.plot(x,y)
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.title("pressure/time")

plt.xlabel("s")
plt.ylabel("MPa")

plt.grid()
plt.plot(x,y)
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.title("pressure/time")

plt.xlabel("s")
plt.ylabel("MPa")

plt.grid()
plt.plot(x,y,marker="o") # o,*
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.title("pressure/time")

plt.xlabel("s")
plt.ylabel("MPa")

plt.grid()
plt.plot(x,y,marker="o",linestyle="--") # :, -, --, -.
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.title("pressure/time")

plt.xlabel("s")
plt.ylabel("MPa")

plt.grid()
plt.plot(x,y,marker="o",linestyle="--",color="green") # red,black,green,...
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [10,20,30,40,50,60]

plt.title("pressure/time")

plt.xlabel("s")
plt.ylabel("MPa")

plt.grid()
plt.plot(x,y,marker="o",linestyle="--",color="green",linewidth=3) 
plt.show()
```
